```python
from google.transit import gtfs_realtime_pb2
import requests
import time
import folium
```

```python
#def get_feed():
pb_list=list()
API_KEY="pg85k9iUkIKKMgQDpHGnN6bsTg0JV7vi"
API_URL = f"https://otd.delhi.gov.in/api/realtime/VehiclePositions.pb?key={API_KEY}"
feed = gtfs_realtime_pb2.FeedMessage()
response = requests.get(API_URL)
```

```python
response.status_code
```

```python
feed.ParseFromString(response.content)
len(feed.entity)
```

```python
print(feed.entity)

for entity in feed.entity:
  vehicle_id=entity.vehicle.vehicle.id
  vehicle_lat=entity.vehicle.position.latitude
  vehicle_lon=entity.vehicle.position.longitude
  vehicle_route_id=entity.vehicle.trip.route_id
#we will create a list of dictionaries 
# every element of the list (array) is a dictionary
# every dictionary is for one bus number plate and its various attributes 
  entity_data={'vehicle_id':vehicle_id,
               'vehicle_lat':vehicle_lat,
               'vehicle_lon':vehicle_lon,
               'vehicle_route_id':vehicle_route_id
               }
  pb_list.append(entity_data)
```

```python
pb_list[0:5]
```

```python
map=folium.Map(location=[28.630,77.217],zoom_start=11)
for vehicle in pb_list:
  lon=vehicle['vehicle_lon']
  lat=vehicle['vehicle_lat']
  marker=folium.CircleMarker(radius=4,location=(lat,lon))
  marker.add_to(map)
map
```
